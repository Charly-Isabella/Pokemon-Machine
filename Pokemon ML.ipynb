```python
file_name = "../input/pokemon/pokemon_alopez247.csv"
pokemon = pd.read_csv(file_name)

pokemon.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

y = pokemon.HP
features = ["Attack","Defense","Sp_Atk","Sp_Def","Speed","Height_m", "Weight_kg"]
X = pokemon[features]

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

#Using decision tree, no max leaf
# Specify Model
first_model = DecisionTreeRegressor(random_state=1)
# Fit Model
first_model.fit(train_X, train_y)
#Predictions
val_predictions = first_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print(val_mae)
```

```python
#Max leaf nodes
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)

#Max leaf
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
#Model with max leaf nodes = 5

second_model = DecisionTreeRegressor(max_leaf_nodes=5, random_state=1)
second_model.fit(train_X, train_y)
val_predictions = second_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print(val_mae)
```

```python
#Model with regression trees
third_model = RandomForestRegressor(random_state=1)
third_model.fit(train_X, train_y)
rf_val_predictions = third_model.predict(val_X)
rf_val_mae = mean_absolute_error(rf_val_predictions, val_y)
print(rf_val_mae)
```

We see that decision trees with max 5 leaves is the most accurate model.
